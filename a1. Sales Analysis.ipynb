#### Q5. Create 2 subsets of the data
##### 1. df_qtr: Calculate sum  units and sales by group ,state, time, year-month, quarter


```python
# Grouping by multiple categories and summing
df_qtr = df_dataonly.groupby(['Group', 'State', 'Time', 'Year-month', 'Quarter'])[['Unit', 'Sales']].sum().reset_index()

# Display the first few rows
print(df_qtr.head())
```

##### 2. df_yr: Calculate sum and average units and sales by date, state, time, week, month, year

```python
# Grouping and calculating multiple statistics
df_yr = df_dataonly.groupby(['Date', 'State', 'Time', 'Week', 'Month', 'Year']).agg({
    'Unit': ['sum', 'mean'],
    'Sales': ['sum', 'mean']
}).reset_index()

# Flattening the MultiIndex columns for easier access
df_yr.columns = ['_'.join(col).strip() if col[1] else col[0] for col in df_yr.columns.values]

print(df_yr.head())
```
